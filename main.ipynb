### Preprocessing

```python
import numpy as np
```

```python
INPUT_PATH = 'input.txt'
MIN_SUPPORT = 0.2  # ex. 25% = 0.25
MIN_CONFIDENCE = 0.2
```

```python
list_trans_set = list()
file = open(INPUT_PATH, 'r')
for line_of_each_trans in file.readlines():
    each_trans_item_set = line_of_each_trans.split()
    list_trans_set.append(each_trans_item_set)
```

```python
print(list_trans_set[1])
```

```python
## Inverted index
from collections import defaultdict

dict_inverted_index_of_item = defaultdict(list)
for idx, item_set in enumerate(list_trans_set):
    for item in item_set:
        dict_inverted_index_of_item[item].append(idx + 1) # trans started index: 1
```

```python
len(dict_inverted_index_of_item) ## item number:16470
```

```python
dict_inverted_index_of_item['0']
```

```python
dict_item_frq = dict()  # support_count
for item, trans_set in dict_inverted_index_of_item.items():
    dict_item_frq[item] = len(trans_set)
```

```python
dict_item_frq
```

### Filter min_support for 1-item

```python
dict_match_min_support = dict()

for item, item_count in dict_item_frq.items():
    if item_count/len(list_trans_set) >= MIN_SUPPORT:
        dict_match_min_support[item] = item_count
```

```python
len(dict_match_min_support)
```

```python
for item, item_count in dict_match_min_support.items():
    print(item, len(dict_inverted_index_of_item[item]))
    break
```

### Find all 2-item

```python
match_min_support_item_set = list()
for item, item_count in dict_match_min_support.items():
    match_min_support_item_set.append(item)
```

```python
len(match_min_support_item_set)
```

```python
match_min_support_item_set
```

```python
# generate all permutations of a list in Python (num = 2)
import itertools
# print(list(itertools.permutations(match_min_support_item_set, 2)))
candidate_item_pair = list(itertools.combinations(match_min_support_item_set, 2))
```

```python
# find match item pair in trans_set
set_match_candidate_in_trans_set = set()
dict_inverted_index_of_2item = defaultdict(list)

for candidate_item in candidate_item_pair:
    subset = set(list(candidate_item))
    for idx, trans in enumerate(list_trans_set):
        outerset = set(trans)
        if subset <= outerset:
            set_match_candidate_in_trans_set.add(candidate_item)
            dict_inverted_index_of_2item[candidate_item].append(idx + 1)
```

```python
dict_inverted_index_of_2item
```

```python
dict_2item_frq = dict()  # support_count
for item, trans_set in dict_inverted_index_of_2item.items():
    dict_2item_frq[item] = len(trans_set)
```

```python
dict_2item_frq
```

### Filter min_support for 2-item

```python
dict_match_min_support_2item = dict()

for item, item_count in dict_2item_frq.items():
    if item_count/len(list_trans_set) >= MIN_SUPPORT:
        dict_match_min_support_2item[item] = item_count
```

```python
len(dict_match_min_support_2item)
```

```python
dict_match_min_support_2item
```

### Find all 3-item

```python
match_min_support_item_set_2item = list()
for item, item_count in dict_match_min_support_2item.items():
    match_min_support_item_set_2item.append(item)
```

```python
match_min_support_item_set_2item
```

```python
candidate_3item_pair = list()
for idx in range(len(match_min_support_item_set_2item)):
    for idx2 in range(idx + 1, len(match_min_support_item_set_2item)):
#         print(idx, idx2)
#         print(len(set(match_min_support_item_set_2item[idx]) | set(match_min_support_item_set_2item[idx2])))
        if len(set(match_min_support_item_set_2item[idx]) | set(match_min_support_item_set_2item[idx2])) == 3:
            candidate_3item_pair.append(tuple(sorted(list(set(match_min_support_item_set_2item[idx]) | set(match_min_support_item_set_2item[idx2])))))
```

```python
match_min_support_3item_set = list()
from collections import Counter

dict_candidate = dict((x,candidate_3item_pair.count(x)) for x in set(candidate_3item_pair))
a = [i for i, j in dict_candidate.items() if j == 3]
for pair in a:
    must_match_subset = list(itertools.combinations(list(pair), 2))
    if set(must_match_subset) <= set(match_min_support_item_set_2item):
        match_min_support_3item_set.append(pair)
```

```python
match_min_support_3item_set
```

```python
# find match item pair in trans_set
set_match_candidate_in_trans_set = set()
dict_inverted_index_of_3item = defaultdict(list)

for candidate_item in match_min_support_3item_set:
    subset = set(list(candidate_item))
    for idx, trans in enumerate(list_trans_set):
        outerset = set(trans)
        if subset <= outerset:
            set_match_candidate_in_trans_set.add(candidate_item)
            dict_inverted_index_of_3item[candidate_item].append(idx + 1)
```

```python
dict_inverted_index_of_3item
```

```python
dict_3item_frq = dict()  # support_count
for item, trans_set in dict_inverted_index_of_3item.items():
    dict_3item_frq[item] = len(trans_set)
```

```python
dict_3item_frq
```

### Filter min_support for 3-item

```python
dict_match_min_support_3item = dict()

for item, item_count in dict_3item_frq.items():
    if item_count/len(list_trans_set) >= MIN_SUPPORT:
        dict_match_min_support_3item[item] = item_count
```

```python
dict_match_min_support_3item
```

### Filter min_confidence

```python
result = ''
```

```python
for item, item_count in dict_match_min_support_2item.items():
    a, b = item
#     print(dict_item_frq[a])
#     print(dict_2item_frq[item])
    if dict_item_frq[a] / dict_2item_frq[item] >= MIN_CONFIDENCE:
        print(a + "->" + b)
        result += a + "->" + b + '\n'
```

```python
for item, item_count in dict_match_min_support_3item.items():
    a, b, c = item
    if dict_item_frq[a] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(a + "->" + b + ' ' + c)
        result += a + "->" + b + ' ' + c + '\n'
    if dict_item_frq[b] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(b + "->" + a + ' ' + c)
        result += b + "->" + a + ' ' + c + '\n'
    if dict_item_frq[c] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(c + "->" + a + ' ' + b)
        result += c + "->" + a + ' ' + b + '\n'
    if dict_item_frq[a] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(a + "->" + b + ' ' + c)
        result += a + "->" + b + ' ' + c + '\n'
    if dict_2item_frq[tuple([a, b])] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(a + ' ' + b + "->" + c)
        result += a + ' ' + b + "->" + c + '\n'
    if dict_2item_frq[tuple([b, c])] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(b + ' ' + c + "->" + a)
        result += b + ' ' + c + "->" + a + '\n'
    if dict_2item_frq[tuple([a, c])] / dict_3item_frq[item] >= MIN_CONFIDENCE:
        print(a + ' ' + c + "->" + b)
        result += a + ' ' + c + "->" + b + '\n'
```
